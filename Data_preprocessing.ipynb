# Data preprocessing Steps
* Import basic libraries.
* Read excel file using pandas.
* Check out unique values.
* Check for the missing values and handle them.
* Encode the categorical data using Label Encoder or One hot encoding.
* Scale the data using Standard scaler or Min max scaler.


```python
# Import basic Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# read data file
data=pd.read_excel('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls')
data.head()
```

```python
# print shape of the data
data.shape
```

```python
# check for datatypes
data.info()
```

```python
# print columns
data.columns
```

```python
# Droping irrelevent column like EmpNumber
data=data.drop('EmpNumber',axis=1)
data.head(2)
```

```python
# print columns which contain numerical data
num_data=data.select_dtypes(include=['int64','float64'])
num_data.columns
```

```python
# Print columns which contain categorical data
cat_data=data.select_dtypes(include=['object'])
cat_data.columns
```

```python
# check for unique values in data
for x in cat_data:
 print(x)
 print(data[x].unique())
 print(data[x].value_counts())
```

```python
# check for missing values or nan values
data.isnull().sum().any()
```

```python
# check for duplicates
data.duplicated().any()
```

## There are no missing values and duplicates in the given dataset.

## Using One Hot Encoding method to handle categorical data parameters.
- This dataset contains categorical columns(string values) where machine learning model cannot work on such data. So we convert them into numerical data using One hot encoder or label encoder.

```python
new_data=pd.get_dummies(data)
new_data.shape
```

## Apply Min max scaler to scale the data

```python
# Split data into X and Y
X=new_data.drop('PerformanceRating',axis=1)
Y=new_data[['PerformanceRating']]
```

```python
#Apply min max scaler
from sklearn.preprocessing import MinMaxScaler
model_max=MinMaxScaler()
print(model_max.fit(X))
print(model_max.data_max_)
```

```python
data2=model_max.transform(X)
data2
```

```python
scaled_data=pd.DataFrame(data2, columns = X.columns)
scaled_data.head(4)
```

```python
X=scaled_data
Y=new_data[['PerformanceRating']]
```

```python
new_data1=X.join(Y)
new_data1
```

# Save transformed data as excel file.

```python
new_data1.to_excel('Employee_analysis_transformed_data.xls')
```
